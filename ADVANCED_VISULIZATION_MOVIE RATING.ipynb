```python
# Basically joint plot is a scatter plot & it find the relation between audience & critics
# Also if you look up you can find the uniform distubution(critics) and normal distrubution(audience)
```

```python
j = sns.jointplot(data = movies, x = 'CriticRatings', y = 'AudienceRatings')
plt.show()

# Audience rating is more dominant then critics ratings
# Based on this we find out as most people are not likelihood to watch audience rating & less likely to watch critics rating
# there is positive corelation between 2 attributes
# let me explain the excel - if you filter audience rating &critic rating has very low
```

```python
j = sns.jointplot(data = movies, x = 'CriticRatings', y = 'AudienceRatings', kind='hex')
plt.show() 
```
